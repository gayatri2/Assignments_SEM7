## ML 4 (To find local minima using gradient descent, start from x=2 for y=(x+3)^2)

```python
cur_x=2
rate=0.01
precision=0.000001
prev_step_size=1
iters=0
max_iters=10000
df=lambda x: 2*(x+3)
```

```python
while prev_step_size > precision and iters < max_iters:
    prev_x=cur_x
    cur_x=cur_x-rate*df(prev_x)
    prev_step_size=abs(cur_x-prev_x)
    iters=iters+1
    print(iters, " ",cur_x)
```

```python
print(cur_x)
```

This is the local minima.

## ML 2 (KNN and SVM, emails.csv)

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
df=pd.read_csv('emails.csv')
df.head()
```

```python
x=df.iloc[:,1:3001]
y=df.iloc[:,-1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
knn=KNeighborsClassifier(n_neighbors=8)
knn.fit(x_train, y_train)
y_pred=knn.predict(x_test)
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test, y_pred))
```

```python
print(metrics.r2_score(y_test, y_pred))
```

```python
svc=SVC(C=1.0, kernel='rbf', gamma='auto')
svc.fit(x_train, y_train)
y_pred=svc.predict(x_test)
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test, y_pred))
```

```python
print(metrics.r2_score(y_test, y_pred))
```

```python
sns.boxplot(x='fare_amount', data=df)
```

```python
Q1 = np.percentile(df['fare_amount'], 25, interpolation='midpoint')
Q3 = np.percentile(df['fare_amount'], 75, interpolation='midpoint')
IQR=Q3-Q1
upper=Q3+(IQR*1.5)
lower=Q1-(IQR*1.5)
```

```python
df[(df['fare_amount'] < lower)| (df['fare_amount'] > upper)][{'fare_amount'}]
```

```python
corr_matrix=round(df.corr(),2)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
ss = preprocessing.StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

```python
df.head()
```

```python
lrm = LinearRegression()
lrm.fit(x_train, y_train)
y_pred = lrm.predict(x_test)
```

```python
print(lrm.coef_)
```

```python
print(metrics.r2_score(y_test, y_pred))
```

```python
rfr = RandomForestRegressor(n_estimators=10, random_state=1)
rfr.fit(x_train, y_train)
y_pred = rfr.predict(x_test)
```
