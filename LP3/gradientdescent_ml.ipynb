```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
cur_x=2
prev_step_size=1
iters=0
max_iters=10000
precision=0.000001
rate=0.01
df=lambda x: 2*(x+3)
```

```python
while prev_step_size > precision and iters < max_iters:
    prev_x = cur_x
    cur_x = cur_x - rate*df(prev_x)
    prev_step_size = abs(cur_x - prev_x)
    iters = iters + 1
    print(iters , " ", cur_x)
```
